```python
#3. How many different unique email domains names are there in the data set?
emails = df_nyc['company email'].dropna()
emails
```

```python
# 4. Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?

#df_nyc = pd.read_csv(fpath + fname)
df_nyc = pd.read_csv(fname)
df_nyc['NTA'].value_counts()
```

```python
# this returns the number of companies within each NTA
counts = df_nyc[['Company Name','NTA']].groupby('NTA').count()
counts
```

```python
check = counts[counts['Company Name'].ge(5)]
check['Company Name'].sum()
```

```python
# collect all ntas with at least 5 listed businesses:
ntas = counts[counts['Company Name'].ge(5)].index
ntas = list(ntas)
len(ntas)
```

```python
# write a fxn to check if value is in ntas:

def check_nta(nta,ntas):
    tf = bool()
    if nta in ntas:
        tf = True
    else:
        tf = False
    return tf

#df_nyc.loc[df_nyc['NTA'] in ntas]
ntas_5 = df_nyc[df_nyc['NTA'].apply(check_nta,args = (ntas,))]
ntas_5
```

```python
ntas_5 = ntas_5[['Total Savings','Job created','NTA']]

#Total Savings is a string with commas - remove the commas and convert to a number

ntas_5 = ntas_5.assign(Total_Savings = ntas_5['Total Savings'].apply(lambda x: float(x.replace(",",""))))
ntas_5.info()
```

```python
jc = ntas_5.groupby('NTA').sum()['Job created']
jc
```

```python
avgts = ntas_5.groupby('NTA').mean()['Total_Savings']
avgts = pd.DataFrame(avgts)
avgts
```

```python
df_final = avgts.assign(Total_Jobs_Created = jc)
df_final = df_final.rename(columns = {'Total_Savings': 'Avg_Total_Savings'})
df_final
```

```python
# 5. Save your result for the previous question as a CSV file.
#df_final.to_csv(fpath + 'Day4_Q5.csv', index = True)
df_final.to_csv('Day4_Q5.csv', index = True)
```

```python
# line plot of the total jobs created for each month
jobs = df_nyc[['Job created','Effective Date']]
jobs.head()
```

```python
# convert effective date to time series:
jobs = jobs.assign(Effective_Date = pd.to_datetime(jobs["Effective Date"]))
jobs.head()
```

```python
jobs['Effective_Date'].value_counts()
```

```python
#groupby effective date and sum jobs for total:
grouped_jobs = jobs.groupby('Effective_Date').sum()
grouped_jobs
```

```python
grouped_jobs.plot()
plt.ylabel('Total Jobs Created');
```

```python
# above is not accurate because it does not account for every month and is not summed within month: need to resample
monthly_jobs = grouped_jobs.resample('M').sum()
monthly_jobs
```

```python
monthly_jobs = monthly_jobs.rename(columns = {'Job Created': 'Jobs Created'})
monthly_jobs.plot()
plt.ylabel('Total Number of Jobs Created');
```
